```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/franklin-univ-data-science/data/master/credit_default.csv')
```

Correlation Heatmap

This type of plot alows the user to identify highly correlated variables and allows them to streamline the feature selection process.

```
cor = df.corr()
plt.figure(figsize=(18,18))
sns.heatmap(cor, cbar = True,  square = True, annot=True, fmt= '.2f',annot_kws={'size': 10},
            xticklabels=cor.columns.values,
            yticklabels=cor.columns.values)
```

Plot data and regression model fits across a FacetGrid.

The FacetGrid class is useful when you want to visualize the distribution of a variable or the relationship between multiple variables separately within subsets of your dataset.

```
sns.lmplot(x='X1', y='Y', data=df)
```

Boxplot

This type of graph is used to show the shape of the distribution, its central value, and its variability

```
sns.boxplot(data=df)
```
