## About the assignment

In the industry, analytics professionals often explore data's measure variables grouped by one or multiple dimension variables. For example, the metrics of interest could be the average, max, min, or percentage (mix) of sales by month, location, etc.

The followings are the sample code snippet using the custom user functions together with `groupby()` and `apply()` in another way:


```
import pandas as pd

# the mpg data is used for demonstration
df = pd.read_csv("https://raw.githubusercontent.com/franklin-univ-data-science/data/master/mpg.csv")
df.head()
```

```
def summarize_data(g):
  return(
    pd.Series({
      "avg_MPG": g.mpg.mean(),
      "min_MPG": g.mpg.min(),
      "max_MPG": g.mpg.max(),
      "sample_count": len(g),
      "mix": g.cylinders.sum()/df.cylinders.sum(),
    })
  )

df.groupby("cylinders").apply(summarize_data)
```

```
df = pd.read_csv('https://raw.githubusercontent.com/frankData612/data_612/master/State_Drug_Utilization_Data_2010/State_Drug_Utilization_Data_2010.csv').dropna()


df
```

```
def summarize_data(g):
  return(
    pd.Series({
      "avg Number of Prescriptions": g['Number of Prescriptions'].mean(),
      "avg Total Amount Reimbursed": g['Total Amount Reimbursed'].mean(),
      "max Total Amount Reimbursed": g['Total Amount Reimbursed'].max(),
      "sample_count": len(g),
      "Percent of Prescriptions": 100*g['Number of Prescriptions'].sum()/df['Number of Prescriptions'].sum(),
    })
  )

df_result = df.groupby(["Quarter","State"]).apply(summarize_data)
df_result.sort_values(by=['Quarter','Percent of Prescriptions'], ascending = False).head(51)
```
